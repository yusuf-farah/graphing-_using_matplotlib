```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Graph

```python
#plt.plot(x, y, label ='2x', color='red', linewidth=2, marker='.', linestyle='--',markersize=12,markeredgecolor='blue')
#use shorthand notation
# format = '[color][marker][line]'

x = [0,1,2,3,4]
y = [0,2,4,6,8]
#resize the graph
plt.figure(figsize=(5,3), dpi=100)


plt.plot(x, y,'r^--',label ='2x')

# line number 2
x2 = np.arange(0,4.5, 0.5)
plt.plot(x2, x2**2, label='x^2')

plt.title('Our first Graph',fontdict={'fontname':'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])
plt.savefig('my graph.png', dpi=300)

plt.legend()
plt.show()
```

# Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
bars = plt.bar( labels, values)
pattern = ['/', 'o','*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')


plt.figure(figsize=(6,4))


plt.show()
```

# Gas Price data from different countries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.head()
```

# Line Graph

```python

plt.figure(figsize = (8,5))
plt.title('Gas Price over Time (in USD)', fontdict = {'fontweight':'bold', 'fontsize':15})


plt.plot(gas.Year, gas.USA, 'b.-', label = 'United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label = 'Canada')
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label = 'South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label = 'Australia')

# other way to graph
#coutries_to_look_at = ['USA', 'Australia', 'Canada', 'South Korea']
#for country in gas:
    #if country in coutries_to_look_at:
        #plt.plot(gas.Year, gas[country], marker='.')
    
    
#print(gas.Year[::3])

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.savefig('Gas_price.png', dpi=300)

plt.legend()


plt.show()
```

# Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#429cf5')

plt.xticks(bins)

plt.xlabel('Skill Level')

plt.ylabel('Number of Players')

plt.title('Distribution of Players Skills in FIFA 2018')


plt.show()
```

```python
left =  fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right =  fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
```

```python
right
```

```python
labels = ['left', 'right']
colors = ['#36a832', '#3271a8']
plt.pie([left, right], labels=labels, colors = colors, autopct='%.2f%%')
plt.title('Foot Preference of FIFA Players')


plt.show()
```
